### 2. Using min_max Model

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
X_scaled = min_max_scaler.fit_transform(X)
```

## Spliting the Dataset into two: train and test

The dataset is split into portion i.e. the train data and the test data. The train data is used to our machine and help in learning the categories where as test data is not shown to the machine in order to check whether the machine is working accurately or not. 

There is one more test dataset available in the repo. You can use that dataset to check the accuracy of the model. Here, I have choosen to split the same data set into two in a ratio of 80%-20%.

```python
from sklearn.model_selection import train_test_split
```

```python
#Split the dataset into two (80-20 ratio) for Training and Testing respectively
X_train,X_test,y_train,y_test=train_test_split(X_scaled, y, test_size=0.2,random_state=42,stratify=y)
```

## Training the SVM model on the training set

```python
#Import SVM Classifier
from sklearn import svm
```

```python
#Create the model
model=svm.SVC(kernel='rbf')
```

```python
#Train the Model
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```

### Checking accuracy of result

```python
from sklearn import metrics
```

```python
#Checking the accuracy of the model
accuracy=metrics.accuracy_score(y_test,y_predict)
accuracy
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_predict))
```
